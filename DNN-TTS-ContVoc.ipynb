# Read and plot the synthesized wave file

```python
%cd /home/malradhi/AI4EU/jupyter/merlin/egs/slt_arctic/s1/experiments/slt_arctic_full/test_synthesis/wav

import IPython
from scipy.io import wavfile as wav
import matplotlib.pyplot as plt
import librosa

wav_input = "test_tts_synthesized_with_Hilbert.wav"
fs, data = wav.read(wav_input)
plt.plot(data)
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.show()

IPython.display.Audio(wav_input, rate=fs, autoplay=True)
```
